```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import imageio
from scipy import ndimage
```

```python
image = imageio.imread('duck.jpg')
```

```python
image = image/255
```

```python
plt.imshow(image)
```

```python
#Sharpen
a = np.array([0, -1, 0, -1, 5, -1, 0, -1, 0]).reshape((3,3,1))
#Box blur
#a = np.array([1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9, 1/9]).reshape((3,3,1))
```

```python
image_convolved = ndimage.convolve(image[:,:,:], a, mode='constant', cval=0.0)
```

```python
plt.imshow(image_convolved)
```
